```python
import nltk
```

```python
from konlpy.corpus import kobill
files_ko = kobill.fileids()
doc_ko = kobill.open('1809890.txt').read()
doc_ko
```

```python
from konlpy.tag import Okt
okt = Okt()
tokens_ko = okt.nouns(doc_ko)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name='대한민국 국회 의안 제 1809890호')
print(len(ko.tokens))       # returns number of tokens (document length)
print(len(set(ko.tokens)))  # returns number of unique tokens
ko.vocab()                  # returns frequency distribution
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
ko.plot(50)
plt.show()
```

```python
stop_words = ['의', '자', '에', '안', '번', '호', '을', '이', '다', '만', '로', '가', '를', '발', '인', '김', '태', '완'
            , '및', '장', '정', '문', '종', '생', '략']
ko = [word for word in ko if word not in stop_words]
ko
```

```python
ko = nltk.Text(ko, name='대한민국 국회 의안 제 1809890호')
plt.figure(figsize=(12, 6))
ko.plot(50)
plt.show()
```

```python
ko.count('초등학교')
```

```python
plt.figure(figsize=(12, 6))
ko.dispersion_plot(['육아휴직', '초등학교', '공무원'])
```

```python
# 원하는 단어의 문서 내 위치를 개략적으로 그 분량과 함께 알아보기
# 원하는 단어의 주변부 단어까지 확인 가능
ko.concordance('초등학교')
```

```python
# 문서 내에서 어떤 단어들이 연어(collocation)로 사용되었는지?
# nltk.download('stopwords')
ko.collocations()
```

```python
from wordcloud import WordCloud
data = ko.vocab().most_common(150)

wc = WordCloud(font_path='./resources/Goyang.ttf',
                relative_scaling=0.2,
                background_color='white',
                ).generate_from_frequencies(dict(data))

plt.figure(figsize=(12,8))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
