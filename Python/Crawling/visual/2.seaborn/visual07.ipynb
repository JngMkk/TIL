```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('penguins')
```

```python
"""
species : 종
island : 서식지
bill_length_mm : 부리의 길이
bill_depth_mm : 부리의 깊이
flipper_length_mm : 날개의 길이
body_mass_g : 체질량
sex : 성별
"""

sns.scatterplot(data=data, x='body_mass_g', y='flipper_length_mm')
plt.show()
```

```python
sns.scatterplot(data=data, x='body_mass_g', y='flipper_length_mm', hue='sex')
plt.show()
```

```python
sns.scatterplot(data=data, x='body_mass_g', y='flipper_length_mm', hue='species', size='sex')
plt.show()
```
