```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('penguins')
```

```python
"""
species : 종
island : 서식지
bill_length_mm : 부리의 길이
bill_depth_mm : 부리의 깊이
flipper_length_mm : 날개의 길이
body_mass_g : 체질량
sex : 성별
"""

fig = plt.figure(figsize=(10,7))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 4)
sns.histplot(data=data, x='body_mass_g', ax=ax1)
sns.scatterplot(data=data, x='bill_length_mm', y='bill_depth_mm', ax=ax2)
ax3.boxplot(data['body_mass_g'].fillna(data['body_mass_g'].mean()))
plt.tight_layout()
plt.show()
```
