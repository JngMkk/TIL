```python
from wordcloud import WordCloud
import json
```

```python
cloud = WordCloud(background_color='white', max_words=30, width=400, height=300, font_path='./font/Goyang.ttf')

with open('../../Naver/webtoons.json') as f:
    data = json.load(f)
data
```

```python
d = {}
for elem in data['webtoons']:
    d[elem['title']] = int(float(elem['rating']) * 100)

visual = cloud.fit_words(d)
visual.to_image()
```
