```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import json
```

```python
with open('../../Naver/webtoons.json') as f:
    data = json.load(f)
```

```python
d = {}
for elem in data['webtoons']:
    d[elem['title']] = int(float(elem['rating']) * 100)

masking_img = np.array(Image.open('./images/kakao.png'))
cloud = WordCloud(font_path='./font/Goyang.ttf', max_font_size=40, mask=masking_img, background_color='white').fit_words(d)
cloud.to_file('./images/cloud02.png')
plt.imshow(cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
