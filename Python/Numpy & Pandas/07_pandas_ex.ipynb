### 연습문제2. : 수치형-> 범주형 데이터 생성

: 타이타닉호 승객을 사망자와 생존자 그룹으로 나누고(alive), 각 그룹에 대해 미성년자, 청년, 중년, 장년, 노년 승객의 비율을 구하시오.

- bins=[1,15,25,35,60,99]
- labels=['미성년자','청년','중년','장년','노년']
- 각 그룹별 비유의 전체 합은 1이 되어야 한다.

**1. 타이타닉 데이터셋 로딩**

```python
# 타이타닉 승객 dataset 읽어오기
import seaborn as sns 
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.info()
```

```python
titanic.count()
```

**2. 데이터 셋 분리 : 생존자 그룹과 사망자 그룹**

```python
# 사망자 그룹 데이터 프레임 추출
df2 = titanic[titanic['alive']=='no']
df2.reset_index(drop=True, inplace=True)
```

```python
# 생존자 그룹 데이터 프레임 추출
df1 = titanic[titanic['alive'] == 'yes']
df1.reset_index(drop=True, inplace=True)
```

**3. 범주형 데이터 변경하기 위한 범위 설정**

```python
# 라벨과 범위 간격 설정
labels = ['미성년자', '청년', '중년', '장년', '노년']
bins = [0, 15, 25, 35, 60, 99]
```

**4. 범주형 데이터 생성**

```python
# 카테고리 데이터 생성 : 생존자 그룹, 사망자그룹 별로 생성
df1['age_cut'] = pd.cut(df1.age, bins, labels=labels)
df2['age_cut'] = pd.cut(df2.age, bins, labels=labels)
```

**범주형 데이터 확인**

```python
#카테고리 데이터 확인
df1
df2
```

**5. 범주형 데이터 빈도 계산**

- 사망자 그룹의 연령대별 승객수

```python
# 카테고리 데이터들의 빈도 수 계산
df2['age_cut'].value_counts()

# 사망자그룹 계산
df2['age_cut'].count()

# 각 연령대별 승객수의 비율 계산
p = df2['age_cut'].value_counts() / df2['age_cut'].count()
p

# 전체 비율의 합 계산
sum(p)
```

- 생존자 그룹 빈도

```python
# 생존자그룹 계산
df1['age_cut'].value_counts()
df1['age_cut'].count()
```

- 생존자 그룹의 연령대별 빈도

```python
p = df1['age_cut'].value_counts() / df1['age_cut'].count()
p
sum(p)
```
