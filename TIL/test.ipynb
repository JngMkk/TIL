```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from folium import Marker

df = pd.read_csv('lift.csv', index_col='구', encoding='utf-8-sig')
df = df.dropna(how='all')
df = df.dropna(how='all', axis=1)
for i, l in enumerate(df['위도']):
    df['위도'][i] = l.replace(',', '')
lift = df.iloc[:,[1, 5, 6]]

lift_map = folium.Map(location=[37.55, 126.98], zoom_start = 11)
mc = MarkerCluster()
for _, row in lift.iterrows():
    mc.add_child(
        Marker(location = [row['위도'], row['경도']], tooltip = f"승강기 개수 : {row['승강기 개수']}")
    )
lift_map.add_child(mc)
lift_map.save('lift.html')
```

```python
import pandas as pd
import folium
pd.set_option('display.max_rows', 100)
df = pd.read_csv('lift.csv',encoding='utf-8-sig')
df = df.dropna(how='all')
df = df.dropna(how='all', axis=1)
# df['에스컬레이터 개수'] = df['에스컬레이터 개수'].astype('int')
# df['휠체어리프트 개수'] = df['휠체어리프트 개수'].astype('int')
# df['전동휠체어 급속충전기 개수'] = df['전동휠체어 급속충전기 개수'].astype('int')
# df['승강기 개수'] = df['승강기 개수'].astype('int')
df.drop([0], inplace=True)
df.reset_index(inplace=True, drop=True)
df
```

```python
group_df = df.groupby(df['구'])
total_df = group_df.sum()
total_df['비율'] = total_df['승강기 개수'] / total_df['경도']
total_df['비율'].mean()
```

```python
import folium
state_geo = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'

m = folium.Map(location=[37.562225, 126.978555], tiles="OpenStreetMap", zoom_start=11)

folium.Choropleth(
    geo_data = state_geo,
    data = total_df['에스컬레이터 개수'],
    key_on = 'feature.properties.name',
    fill_color = 'BuPu',
).add_to(m)
m
```

```python
import pandas as pd

data = pd.read_csv('./student_data.csv', encoding='cp949')
data = data.loc[26:29]
del data['가정별']
data.reset_index(drop=True, inplace=True)
data
```

```python
data.rename(columns={'인구종류별':'학령인구'}, inplace=True)
data
```

```python
data['학령인구'][0] = '초등학생'
data['학령인구'][1] = '중학생'
data['학령인구'][2] = '고등학생'
data['학령인구'][3] = '대학생'
data
```

```python
import matplotlib.pyplot as plt
data.set_index('학령인구', inplace=True)
```

```python
data = data.transpose()
data
```

```python
plt.style.use('ggplot')
ax = data.plot(stacked=True, kind='bar', figsize = (14,10), rot='horizontal')
for rect in ax.patches:
    height = rect.get_height()
    width = rect.get_width()
    x = rect.get_x()
    y = rect.get_y()
    label_text = f'{int(height)}'
    label_x = x + width / 2
    label_y = y + height / 2
    if height > 0:
        ax.text(label_x, label_y, label_text, ha = 'center', va = 'center', fontsize = 12)

ax.legend(bbox_to_anchor=(1.0, 1), loc='upper left', borderaxespad=0., fontsize=10) 
ax.set_title('학령인구', fontsize = 20)
ax.set_xlabel("연도", fontsize = 16)
plt.savefig('./학령인구.png')
```
